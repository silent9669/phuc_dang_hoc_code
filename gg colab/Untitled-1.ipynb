```python
p = 61996545789950044511807056126248141945304434838920379595334089328489467238193
a1_v = 60063441005095171597098858010795632629468865189663817083811738261468640425838
b1_v = 18190931364402322785572512418893267163264663934344191594397727140941016963494
b2_v = 22576394155028824954295303146749596176130404633621215135608577696849350286187
alpha = ((b2_v - b1_v) * inverse_mod(a1_v, p)) % p
R.<x> = GF(p)[]; F.<z2> = GF(p^2, modulus=x^2 - alpha*x + 5)
mu1 = F(a1_v)*z2 + F(b1_v)
p2m1 = p^2 - 1; m_val = p2m1 // 1337
d_exp = ZZ( (Mod(1337, m_val)^31337)^(-1) )
lam = mu1^d_exp
g = z2 + 1; zeta = g^((p^2-1)//1337)
while zeta==1 or zeta^7==1 or zeta^191==1: g+=1; zeta=g^((p^2-1)//1337)
for k in range(1337):
    tr_int = Integer((lam + lam^p).polynomial()[0])
    if tr_int < 2^130: lam_c=lam; print(f"k={k}, tr={tr_int}"); break
    lam *= zeta
det_M = Integer((lam_c*lam_c^p).polynomial()[0])
tr_M  = Integer((lam_c+lam_c^p).polynomial()[0])
s_e   = Integer(((mu1+mu1^p).polynomial()[0]))
s_em1 = Integer(((lam_c^(1337^31337-1)+(lam_c^p)^(1337^31337-1)).polynomial()[0]))
print(f"tr_M={tr_M}, det_M={det_M}")
# ---- IntegerLattice CVP ----
from sage.modules.free_module_integer import IntegerLattice
C00p, C01p = 504431219805193233494638855346643150, 437561477922777662981107858757923689
W = 10^41; min_b = 32*(256^16-1)//255
d_c = [(pow(256,i,p)*s_e)%p for i in range(16)]
SCALE = 2^252
K = 16
B = matrix(ZZ, K+1, K+2)
for i in range(K): B[i,i]=SCALE; B[i,K+1]=d_c[i]
B[K,K+1]=p
L = IntegerLattice(B)
# Find b
zeros = [0] * (K+1)
Tb = (C01p*W - min_b*s_e)%p
vb = L.closest_vector(vector(ZZ, zeros + [Tb]))
c_b = [Integer(vb[i])//SCALE for i in range(K)]
b_val = min_b + sum((c+47)*256^i for i,c in enumerate(c_b))
b_str = ''.join(chr((b_val>>(8*(15-i)))&0xFF) for i in range(16))
print(f"b='{b_str}'")
# Find a
Ta = (C00p*W + det_M*s_em1 - min_b*s_e)%p
va = L.closest_vector(vector(ZZ, zeros + [Ta]))
c_a = [Integer(va[i])//SCALE for i in range(K)]
a_val = min_b + sum((c+47)*256^i for i,c in enumerate(c_a))
a_str = ''.join(chr((a_val>>(8*(15-i)))&0xFF) for i in range(16))
print(f"a='{a_str}'")
d_val = tr_M - a_val
c_val = (a_val*d_val - det_M)//b_val
d_str = ''.join(chr((d_val>>(8*(15-i)))&0xFF) for i in range(16))
c_str = ''.join(chr((c_val>>(8*(15-i)))&0xFF) for i in range(16))
flag = f"BKISC{{{a_str}{b_str}{c_str}{d_str}}}"
print(f"\n{flag}")
```
